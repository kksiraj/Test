# NumPy Vectors and Matrices

Step-by-step notes with code for learning vectors and matrix operations.

## Import NumPy

```python
import numpy as np
```

## Vectors
A vector represents magnitude and direction.

```python
v = np.array([1, 2, 3])
u = np.array([4, 5, 6])

print('v =', v)
print('u =', u)
```

### Basic Operations

```python
print('Addition:', v + u)
print('Subtraction:', v - u)
print('Scalar multiplication:', 2*v)
print('Element-wise multiplication:', v*u)
```

### Dot Product

```python
print('Dot product:', v @ u)
```

### Cross Product

```python
print('Cross product:', np.cross(v, u))
```

### Magnitude and Unit Vector

```python
mag = np.linalg.norm(v)
unit = v / mag

print('Magnitude:', mag)
print('Unit vector:', unit)
```

## Matrices

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print('A =\n', A)
print('B =\n', B)
```

### Matrix Operations

```python
print('A+B =\n', A+B)
print('A-B =\n', A-B)
print('2A =\n', 2*A)
print('Element-wise A*B =\n', A*B)
```

### Matrix Multiplication

```python
print('A @ B =\n', A @ B)
```

### Transpose

```python
print('A.T =\n', A.T)
```

### Determinant

```python
print('det(A) =', np.linalg.det(A))
```

### Inverse

```python
A_inv = np.linalg.inv(A)
print('A inverse =\n', A_inv)
```

### Verify Inverse

```python
print('A @ A_inv =\n', A @ A_inv)
```

## Solving Linear System
Solve:
2x + y = 5
x + 3y = 6

```python
A = np.array([[2,1],[1,3]])
b = np.array([5,6])

x = np.linalg.solve(A,b)
print('Solution:', x)
```

### Verify Solution

```python
print('A @ x =', A @ x)
print('b =', b)
```

## Eigenvalues and Eigenvectors

```python
vals, vecs = np.linalg.eig(A)
print('Eigenvalues:', vals)
print('Eigenvectors:\n', vecs)
```

## Physics Connections

### Work Done (Dot Product)

```python
F = np.array([10,0])
d = np.array([5,2])
print('Work:', F @ d)
```

### Torque (Cross Product)

```python
r = np.array([2,0,0])
F = np.array([0,5,0])
print('Torque:', np.cross(r,F))
```
