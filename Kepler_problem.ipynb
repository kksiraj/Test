$\textbf{Kepler Problem}$

The differnetial equation governing the motion is




$$m\frac{d^2\vec r}{dt^2}=-\frac{k}{r^2}\hat r$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(r,v,t):
    return -r/(np.linalg.norm(r))**3
```

```python
def g(r,v,t):
    return v
```

```python
r=np.array([1,0])
v=np.array([0,1.2])
t=0
dt=0.001
```

```python
X=[]
Y=[]
T=[]
dA=[]
```

```python
while (t<20):
    da=0.5*np.cross(r,v*dt)
    dA.append(da)
    X.append(r[0])
    Y.append(r[1])
    T.append(t)
    r=r+g(r,v,t)*dt
    v=v+f(r,v,t)*dt
    t+=dt
```

```python
A1=sum(dA[1:15])
A2=sum(dA[11:25])
print(round(A1,5),round(A2,5))
```
