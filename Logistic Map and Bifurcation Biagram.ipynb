

The logistic map is a mathematical function used to model population growth. It is defined by the equation:

$$
x_{n+1} = m x_n  (1 - x_n)
$$




```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(m,x):
    return m*x*(1-x)
```

```python
x=np.linspace(0,1)
plt.plot(x,f(0.8,x))

plt.plot(x,f(1.8,x),color="r")
plt.plot(x,f(2.5,x), color="g")
plt.plot(x,f(3.5,x), color="b")

plt.plot(x,x)
```

```python
n=3.4
plt.plot(x,x)
plt.plot(x,f(n,x))
x0=0.02
for p in range(100):
    plt.plot([x0,x0],[x0,f(n,x0)])
    plt.plot([x0,f(n,x0)],[f(n,x0),f(n,x0)])
    x0=f(n,x0)
```

```python
def g(m,x):
    #return m*x-m*x**2
    return m*(m*x-m*x**2)*(1-m*x+m*x**2)
```

```python
x=np.linspace(0,1)
n=3.4
plt.plot(x,g(n,x))
'''
plt.plot(x,g(2.5,x))
plt.plot(x,g(1.8,x))
plt.plot(x,g(0.8,x), color="r")
plt.plot(x,g(4.2,x), color ="b")
'''
plt.plot(x,x)
x0=0.06

for p in range(201):
    plt.plot([x0,x0],[x0,g(n,x0)])
    plt.plot([x0,g(n,x0)],[g(n,x0),g(n,x0)])
    x0=g(n,x0)
```
