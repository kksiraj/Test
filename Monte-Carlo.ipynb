$\textbf{Monte-Carlo Techniques }$

Monte Carlo simulation is a computational technique used to approximate the behavior of complex systems or processes through repeated random sampling. It's named after the famous Monte Carlo Casino due to its reliance on randomness.

In essence, Monte Carlo simulation involves generating a large number of random samples from known probability distributions to simulate uncertain factors or variables in a system. By running these simulations multiple times, it's possible to analyze the range of possible outcomes and estimate probabilities or expected values for various outcomes.

This technique is widely used in fields such as finance, engineering, physics, and statistics to solve problems that are difficult or impossible to solve analytically. It's particularly useful for modeling systems with randomness, uncertainty, or nonlinear relationships.

Consider estimating the value of a definite integral:
$$ I = \int_{a}^{b} f(x) \, dx $$

Monte Carlo simulation can be utilized to estimate the value of a definite integral by following these steps:


$ \textbf{Generate Random Points}:$ Randomly sample points within the region of interest, typically defined by the interval $[a, b]$ along the x-axis.
    
$ \textbf{Evaluate the Function}:$ For each randomly generated x-coordinate, evaluate the function $f(x)$ to be integrated at that point. This yields the corresponding y-coordinate on the function curve.
    
$ \textbf{Summation}:$ Sum up the function values obtained for all sampled points. This sum approximates the area under the curve of the function within the interval $[a, b]$.
    
$ \textbf{Area Approximation}:$ Multiply the average of the summed function values by the width of the integration interval $(b - a)$. This yields an estimate of the integral:
    
   $$I \approx \frac{1}{N} \sum_{i=1}^{N} f(x_i) \times (b - a) $$
    
where $N$ is the total number of sampled points and $x_i$ represents the randomly sampled x-coordinates.
    
$ \textbf{Increasing Accuracy}:$ The accuracy of the estimation improves as the number of random points $N$ increases, following the law of large numbers.


Monte Carlo simulation for integration is particularly useful when traditional analytical methods are impractical or infeasible, such as for high-dimensional integrals or integrals involving complex functions.

$\textbf{The Monte Carlo simulation estimates the value of $\pi$ using the following steps:}$


 Consider a unit circle centered at the origin within a square with side length 2 units.

 Ratio of areas of square to circle is

 $$
 \frac{A_s}{A_c}=\frac{2^2}{\pi 1^2}$$
 $$\pi=\frac{A_c}{A_s} \times 4$$
     
 Randomly generate points within the square, covering the entire area.
     
 Determine whether each point falls inside the unit circle by calculating its distance from the origin.

 Estimate $\pi$ by taking the ratio of the number of points inside the circle to the total number of points, multiplied by 4. Number of points inside the square and circle will be proportional to the respective areas.

$$ \pi=\frac{N_c}{N_s} \times 4 $$

$\textbf{Simulation of radio active decay}$

Number of Parent Nuclei as a function of time is given by
$$
N(t)=N_0e^{-\lambda t}
$$

$N_0$ is the Number of Nuclei at time $t=0$

$\lambda$ is the disintegration constand and is related to half life as

$$
\lambda=\frac{0.693}{t_{half}}
$$

for simplicity let us take initaial number of nucei to be 1000 and half life 10 units of time, Lets code it up and simulate the decay.

```python
N_0=100000
t=0
t_half=10
lmbda=0.693/t_half
```

```python
N_P=N_0
N_D=0
NP=[]
ND=[]
T=[]
```

```python
while N_P>0:
    ND.append(N_D)
    NP.append(N_P)
    T.append(t)
    d=0
    for i in range(N_P):
        r=np.random.uniform(0,1)
        if r<lmbda:
            d+=1
    N_P-=d
    N_D+=d
    t+=1
T=np.array(T)        
NT=N_0*np.exp(-lmbda*T) 
```

```python
plt.plot(T,NP, color="b")
plt.plot(T,ND, color="g")
plt.scatter(T,NT,color="r")
```
