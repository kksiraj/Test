# Differential Equation Systems for Student Projects

## Lorenz Attractor
$$\frac{dx}{dt}=\sigma(y-x)$$
$$\frac{dy}{dt}=x(\rho-z)-y$$
$$\frac{dz}{dt}=xy-\beta z$$

Parameters:

$\sigma=10$

$\rho=28$

$\beta=8/3$

Initial:
x=1,y=1,z=1


## Rössler Attractor
$$\frac{dx}{dt}=-y-z$$
$$\frac{dy}{dt}=x+ay$$
$$\frac{dz}{dt}=b+z(x-c)$$

a=0.2
b=0.2
c=5.7

Initial:
x=1,y=1,z=1


## Duffing Oscillator
$$\frac{dx}{dt}=v$$
$$\frac{dv}{dt}=\gamma\cos(\omega t)-\delta v-\alpha x-\beta x^3$$

alpha=-1
beta=1
delta=0.2
gamma=0.3
omega=1.2

Initial:
x=0.5
v=0


## Van der Pol Oscillator
$$\frac{dx}{dt}=v$$
$$\frac{dv}{dt}=\mu(1-x^2)v-x$$

$\mu=3$

Initial:
x=1
v=0


## Lotka-Volterra
$$\frac{dx}{dt}=\alpha x-\beta xy$$
$$\frac{dy}{dt}=\delta xy-\gamma y$$

$\alpha=1.1$

$\beta=0.4$

$\delta=0.1$

$\gamma=0.4$

Initial:
x=10
y=5
