$\textbf{Charged Particle in Uniform Magnetic Field}$

The differnetial equation governing the motion is




$$m\frac{d^2\vec r}{dt^2}=q(\vec v \times \vec B)$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(r,v,t):
    return np.cross(v,b)
```

```python
def g(r,v,t):
    return v
```

```python
r=np.array([0,0,0])
v=np.array([1,0,0])
b=np.array([0,-2,0])
dt=0.0001
t=0
```

```python
x=[]
y=[]
z=[]
#q/m=1
```

```python
while (t<10):
    x.append(r[0])
    y.append(r[1])
    z.append(r[2])
    r=r+g(r,v,t)*dt
    v=v+f(r,v,t)*dt
    t+=dt
```

```python
P=plt.axes(projection="3d")
P.plot3D(x,y,z)
plt.show()
```
