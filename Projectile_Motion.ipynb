$\textbf{Projectile Motion}$

The differnetial equations governing the motion are

$$m\frac{d^2y}{dt^2}=-mg-bvv_y$$

$$m\frac{d^2x}{dt^2}=-bvv_x$$

or
$$m\frac{d^2\vec r}{dt^2}=-mg\hat y-bv\vec v$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(r,v,t,b):
    a0=np.array([0,-g0])
    a=a0-b*np.linalg.norm(v)*v
    return a
```

```python
def g(r,v,t):
    return v
```

```python
def projectile(r,v,b,t):
    X=[]
    Y=[]
    
    while (r[1]>=0):
        X.append(r[0])
        Y.append(r[1])
        r=r+g(r,v,t)*dt
        v=v+f(r,v,t,b)*dt
        t+=dt
    return X, Y
```

```python
v0=float(input("Enter Projectile velocity "))
th=float(input("Enter Angle of Projectile in Degrees "))
```

```python
th=np.pi/180*th
dt=0.001
g0=9.8
```

```python
r_i=np.array([0,0])
v_i=np.array([v0*np.cos(th),v0*np.sin(th)])
t_i=0.0
```

```python
X,Y=projectile(r_i,v_i,0,t_i)
XA,YA=projectile(r_i,v_i,.2,t_i)
```

```python
plt.plot(X,Y,color="b")
plt.plot(XA,YA,color="r")
```
